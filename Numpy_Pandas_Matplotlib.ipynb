# Датасет 3 - динамика объёма продаж по годам

```python
df = pd.read_csv('/content/DataSet1_3.csv', ",")
df.info()
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.drop(columns = "Name")
df
```

```python
df['year'] = df['Date'].dt.year
df
```

```python
volume_year = df.groupby(df['year'])
volume_year = volume_year['Volume'].sum()
volume_year
```

```python
volume_year.plot(kind='bar', fontsize=10, figsize=(15, 15))
plt.ylabel('Объём по годам')
plt.legend(labels=volume_year.index)
plt.show()
```

# Датасет 4 - динамика заболеваемости в Африке в 2020 и в 2022 годах




```python
df = pd.read_csv('/content/DataSet1_4.csv', ",")
df.info()
```

```python
df
```

```python
df['Day'] = pd.to_datetime(df['Day'])
df = df.drop(columns = "Year")
df
```

```python
df['Year'] = df['Day'].dt.year
df
```

```python
df.Continent.describe()
```

```python
continents = df.groupby('Continent').get_group('Africa')
for continent in continents:
  if (continent[0] == 'Africa'):
    Africa = continent[1]
Africa
```

```python
Years = Africa.groupby('Year')
for Year in Years:
  if (Year[0] == 2020):
    Y20 = Year[1]
  if (Year[0] == 2022):
    Y22 = Year[1]
Y20
```

```python
G20 = pd.DataFrame(data={'Day':[],'New daily case':[]})
G20['Day'] = Y20['Day']
G20['New daily case'] = Y20['Daily new confirmed cases due to COVID-19 (rolling 7-day average, right-aligned)']
G20
```

```python
G20.dropna()
```

```python
G22 = pd.DataFrame(data={'Day':[],'New daily case':[]})
G22['Day'] = Y22['Day']
G22['New daily case'] = Y22['Daily new confirmed cases due to COVID-19 (rolling 7-day average, right-aligned)']
G22.dropna()
```

```python
plt.figure(figsize=(15, 15))
plt.title("Динамика заболеваемости в 2020 году")
plt.xlabel("День")
plt.ylabel("Количество заболевших")
plt.bar(G20['Day'],G20['New daily case'])
```

```python
plt.figure(figsize=(15, 15))
plt.title("Динамика заболеваемости в 2022 году")
plt.xlabel("День")
plt.ylabel("Количество заболевших")
plt.bar(G22['Day'],G22['New daily case'])
```
